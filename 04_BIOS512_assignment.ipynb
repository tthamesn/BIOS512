Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
# make a bar chart to show turbine count per territory
turbine |>
    mutate(province_territory = fct_lump(province_territory, n = 9) |> # make provinces a factor and lump those with small values
          fct_relevel('Other', after = Inf) |> # create an other category that appears last
          fct_infreq() |> # sort by frequency
          fct_rev()) |> # reverse ascending
    ggplot(aes(y = province_territory)) + geom_bar() # plot
```
