Using the `penguins` dataset, make a scatter chart with `bill_length_mm` on the x-axis and `bill_depth_mm` on the y-axis.

```r
ggplot(penguins, aes(x = bill_length_mm, y = bill_depth_mm)) +
    geom_point(na.rm = TRUE, alpha = 0.5) +
    # add a meaningful title
    ggtitle("Bill length v. bill depth") +
    # adjust theme options for title and axes
    theme(
        plot.title = element_text(color = "grey", size = 14, face = "bold", hjust = 0.5),
        axis.title.x = element_text(size=12),
        axis.title.y = element_text(size=12)) +
    # set axes limits and breaks
    scale_x_continuous(name = "Bill length (mm)", limits = c(30, 60)) +
    scale_y_continuous(name = "Bill depth (mm)", limits = c(10, 25))
```

Make the same plot but *encode* the `species` column with color.

```r
# adjust width of plot
options(repr.plot.res = 250, repr.plot.height = 3, repr.plot.width = 5)

ggplot(penguins, aes(x = bill_length_mm, y = bill_depth_mm, color = species)) +
    geom_point(na.rm = TRUE, alpha = 0.7) +
    # add a meaningful title
    ggtitle("Bill length v. bill depth") +
    # adjust theme options for title and axes
    theme(
        plot.title = element_text(color = "grey", size = 14, face = "bold", hjust = 0.5),
        axis.title.x = element_text(size=12),
        axis.title.y = element_text(size=12)) +
    # set axes limits and breaks
    scale_x_continuous(name = "Bill length (mm)", limits = c(30, 60)) +
    scale_y_continuous(name = "Bill depth (mm)", breaks = seq(10, 25, by = 2), limits = c(12, 22)) +
    # add color brewer palette
    scale_color_brewer(type = "qual", palette = "Set2")
```
